# Matrix modifications

Modifications are lazy-proccessed.

```python
nodiag_mtx = sample_mtx.remove_diagonal(ndiag=1, value=0)
```

```python
plt.rcParams['figure.figsize'] = 10, 10
plt.imshow(nodiag_mtx._mtx[0:100, 0:100], cmap='Reds')
```

If you want to apply the modifications, turn matrix into the array or write the matrix to file.

```python
sample_mtx.remove_diagonal(1, 0).as_array()
```

The first available matrix modification is __removing diagonal__. You specify the number of diagonal (1 is the main diagonal) and the value to fill it.

```python
nodiag_mtx = sample_mtx.remove_diagonal(1, 0).as_array()
```

```python
plt.rcParams['figure.figsize'] = 10, 10
plt.imshow(nodiag_mtx[0:100, 0:100], cmap='Reds')
```

The second modification implemented is __filtering extreme values__. The values above `max_percentile` and below `min_percentile` will be replaced with values of respective percentiles.

```python
filtered_mtx = sample_mtx.filter_extreme(min_percentile=1, max_percentile=99).as_array()
```

```python
plt.rcParams['figure.figsize'] = 10, 10
plt.imshow(filtered_mtx[0:100, 0:100], cmap='Reds')
```

The third modification is __logarithming__ by the given base. To produce adequate matrix, `filter_extreme` should be applied prior to logarithming. After procedure subtraction of minimal value should be applied to make values be non-negative and then convert them to integers.

```python
log_mtx = sample_mtx.filter_extreme().log_transform(base=2).subtract_min().convert_to_int().as_array()
```

```python
plt.rcParams['figure.figsize'] = 10, 10
plt.imshow(log_mtx[0:100, 0:100], cmap='Reds')
```

# Write matrix

To write matrix in cool format, you have to specify output file, size of bins, assembly name, chromosome name and size.

```python
sample_mtx._write_mtx('sample_write.cool', binsize=20000, assembly_name='dm3',
                      ch='chr17', chromsize=20000 * sample_mtx._size)
```

# Matrix conversion

You can convert matrix in several ways:
* cool -> txt(.gz)
* cool -> sparse
* cool -> mr_sparse (for MrTADFinder)
* cool -> h5
* cool -> hic
* txt -> cool
* mtx (np.array) -> txt(.gz)
* mtx -> cool

```python
sample_mtx.convert(input_filename=sample_mtx._mtx, output_filename='sample_mtx.txt',
                   input_format=None, output_format='txt')
```

```python
sample_mtx.convert(input_filename='../example_data/HepG2_NA_NA_1.20000.chr17.cool',
                   output_filename='sample_mtx.txt.gz', input_format='cool', output_format='txt.gz',
                   chr='chr17', resolution=20000, balance=False)
```

# Matrix simulation

You can simulate Hi-C data. First, the segmentation is simulated. TAD sizes and interTAD sizes are sampled from uniform distribution. Interaction matrix is sampled from negative binomial distribution with added __interactions__ of given number and __noise__ of given percent value.

```python
TADselect.simulate_HiC(mtx_size=1000, min_tadsize=3, max_tadsize=20, min_intertadsize=0,
                       max_intertadsize=0, tads_filename='../example_data/test_TADs.txt', mtx_filename='../example_data/test_mtx.cool',
                       kdiag=35, kt=28, kd=2 * 28, krandom=1, knoise=2, c=-0.69, pseudo=1,
                       interactions=int(1000 ** 2 / (100000 * 2)), noise=0, dispersion=0.01)
```
